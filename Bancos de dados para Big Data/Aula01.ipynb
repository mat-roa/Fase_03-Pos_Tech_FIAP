```python
SELECT invoice_and_item_number, date, store_number
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE store_number = '5211'
```

```python
SELECT COUNT(invoice_and_item_number) as transactions,
       date,
       store_number
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE store_number = '5211'
GROUP BY date, store_number
ORDER BY date
##ASC & DESC
```

```python
SELECT date,
       store_name,
       category_name,
       count(distinct invoice_and_item_number) transactions,
       sum(volume_sold_gallons) total_volume_sold_gallons,
       sum(volume_sold_liters) total_volume_sold_liters,
       sum(sale_dollars) total_revenue_sale,
       avg(sale_dollars) average_revenue_sale,
       min(sale_dollars) min_revenue_sale,
       max(sale_dollars) max_revenue_sale,
FROM `bigquery-public-data.iowa_liquor_sales.sales`
GROUP BY date,
         store_name,
         category_name
ORDER BY date
```
