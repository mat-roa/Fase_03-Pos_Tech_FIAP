```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
```

#Grafico de genêro

```python
#Grafico de distribuição de genero
df_genero = pd.read_csv("/content/drive/MyDrive/FIAP/Fase 03/Tech Challenge/Tabelas/genero_populacao.csv", sep=',')
df_genero.replace({1: 'Homens', 2: 'Mulheres'}, inplace=True)

color = ['#ffccd5','#9bedff']
explode = (0.02,0.02)

plt.figure(figsize=(15, 8))
fig, ax = plt.subplots()
ax.pie(df_genero['total_entrevistas'], labels=df_genero['Genero'],autopct='%1.1f%%',
       startangle = 90, colors = color, pctdistance=0.65, explode = explode)

centre_circle = plt.Circle((0,0),0.0,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Distribuição dos entrevistados por gênero", pad=15, color='#333333', weight='bold')
plt.show()
```
